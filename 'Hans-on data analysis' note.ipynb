### 思考：什么是逐块读取？为什么要逐块读取？


'ChunkSize
chunk——块。假设我们现在有一个200Mb数据矩阵，我们对这个矩阵进行计算，而内存上限为210Mb, 一不小心就会超出内存，导致后面的计算通通停掉。这个方法通俗讲就是，一块大蛋糕我一口气吃不下，那就切成n块小蛋糕一点一点吃。因此，我们可以看出，这样的方法节省空间，但是会很耗时。时间复杂度和空间复杂度，鱼和熊掌不可兼得'

——https://blog.csdn.net/Monkeyhou1992/article/details/105115879'

```python
#以"Age"为条件，将年龄在10岁以上和50岁以下的乘客信息显示出来，并将这个数据命名为midage
midage=df[(df['Age']>10)&(df['Age']<50)]
midage
```

```python
#将midage的数据中第100行的"Pclass"和"Sex"的数据显示出来
midage.reset_index(inplace=True)
midage.loc[[100],['Pclass','Sex']]
```

### 思考：reset_index()函数的作用

重置索引，不使用重置索引将会使用之前的索引，导向的.loc[[100]]是原索引的100对应值

```python
#使用loc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来
midage.loc[[100,105,108],['Pclass','Name','Sex']]
```

```python
#使用iloc方法将midage的数据中第100，105，108行的"Pclass"，"Name"和"Sex"的数据显示出来
midage.iloc[[99,104,107],[2,3,4]]
```

### 思考：iloc和loc的异同

都是定向索引，iloc传入整型，loc传入列名

## ③探索性数据分析

```python
#创建一个DataFrame
frame=pd.DataFrame(np.arange(8).reshape(2,4)
                  ,index=[2,1]
                  ,columns=['d','a','b','c']
                  )
frame
```

```python
#将frame根据'c'列的值进行升序排列
frame.sort_values(by='c'
                  ,ascending=True#ascengding=False则是降序，默认升序
                 )
```

```python
#将frame根据索引升序排列
frame.sort_index()
```

```python
#将frame根据表头升序排列
frame.sort_index(
    axis=1#axis=0是纵轴，axis=1是横轴
)
```

```python
#将frame根据索引降序排列
frame.sort_index(ascending=False)
```

```python
#将frame根据列值降序排序
frame.sort_values(by=['b','c'],ascending=False)
```

DataFrame相加

```python
#两个例子相加
frame1_a+frame1_b
```
